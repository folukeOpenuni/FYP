```python
 # evolve the gate
n = 20 #25 
psi = [gate(t)*psi0 for t in np.linspace(0, 1., 2*n)] #1.570796 - 90degree, 2.094395 - 120 degree, 6.283185 - 360
# plotting the states. State evolution during the first hamadard gate is red. 
b.vector_color = [matplotlib.colors.to_rgba('r', alpha=i) for i in np.arange(n)/float(n)]
b.add_states(psi)
b.show()
```

```python
import math
b2 = Bloch()

def pauli_y_gate(degreeOfRotation5):
    b2.clear()
    #degreeOfRotation5 = 6.283185 #360   
    degree = math.radians(360)
    pauli_y = qutip.qip.operations.ry(degreeOfRotation5) #the gate 2.094395, 4.094395, 3.28319

    hamiltonian = Qobj(scipy.linalg.logm(pauli_y.data.todense()), dims=pauli_y.dims) / np.pi * 1.j
    
    psi0 = (basis(2, 0)).unit()#create initial state vector

    # evolve the gate
    n = 50 
    psi = [(1.j*hamiltonian*theta).expm()*psi0 for theta in np.linspace(0, degreeOfRotation5, n)] 
    
    # plotting the states. 
    b2.vector_color = [matplotlib.colors.to_rgba('b', alpha=i) for i in np.arange(n)/float(n)]
    b2.add_states(psi)
    b2.show()
```

```python
import math
b2 = Bloch()

def pauli_x_gate(degreeOfRotation5):
    b2.clear()
    #degreeOfRotation5 = 6.283185 #360   
    degree = math.radians(360)
    pauli_x = qutip.qip.operations.rx(degreeOfRotation5) #the gate 2.094395, 4.094395, 3.28319

    hamiltonian = Qobj(scipy.linalg.logm(pauli_x.data.todense()), dims=pauli_x.dims) / np.pi * 1.j
    
    psi0 = (basis(2, 0)).unit()#create initial state vector

    # evolve the gate
    n = 50 
    psi = [(1.j*hamiltonian*theta).expm()*psi0 for theta in np.linspace(0, degreeOfRotation5, n)] 
    
    # plotting the states. 
    b2.vector_color = [matplotlib.colors.to_rgba('g', alpha=i) for i in np.arange(n)/float(n)]
    b2.add_states(psi)
    b2.show()
```
