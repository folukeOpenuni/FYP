```python
#link to github: https://github.com/folukeOpenuni/FYP
from tkinter import *
import numpy as np
from numpy import sqrt
import scipy
import matplotlib.colors
import scipy

import matplotlib

from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
from matplotlib.figure import Figure

from mpl_toolkits.mplot3d import Axes3D # needs Axes3D object to activate the '3d' projection

from qutip import *
import matplotlib.pyplot as plt
import math
```

```python
b2 = Bloch()

def pauli_y_Gate():
    b2.clear()
    degreeOfRotation3 = math.radians(360)
    degree = 1.570796
    pauli_y = qutip.qip.operations.ry(degreeOfRotation3) #the gate 2.094395, 4.094395, 3.28319

    # the hamilton operator for time evolution
    hamiltonian = Qobj(scipy.linalg.logm(pauli_y.data.todense()), dims=pauli_y.dims) / np.pi * 1.j
    
    psi0 = (basis(2, 0)).unit()#create initial state vector

    # evolve the gate
    n = 20 #25 
    psi = [(-1.j*hamiltonian*theta).expm()*psi0 for theta in np.linspace(0, degreeOfRotation3, n)] 
    
    # plotting the states. 
    b2.vector_color = [matplotlib.colors.to_rgba('b', alpha=i) for i in np.arange(n)/float(n)]
    b2.add_states(psi)
    b2.show()
    
```

```python
pauli_y_Gate()
```
